# Dataset Generator

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta


num_rows = 2500  

# Opciones categóricas
channels = ['Programmatic', 'Paid Search', 'Paid Social', 'Organic']
data_sources = ['Amazon Ad', 'Stack', 'LinkedIn', 'Facebook', 'Google', 'Bing', 'Google Search Ads']
campaigns = ['Seamless', 'Entreprise', 'Optional', 'Fundamental', 'Diverse', 'Mandatory']

# Función para generar fechas aleatorias entre 2023 y 2024
def random_date(start_year=2023, end_year=2024):
    start = datetime(start_year, 1, 1)
    end = datetime(end_year, 12, 31)
    return start + timedelta(days=random.randint(0, (end - start).days))

# Generación del dataset
data = []
for _ in range(num_rows):
    impressions = random.randint(1000, 1000000)
    ctr = round(np.random.uniform(0.5, 5.0), 2)  # Porcentaje
    clicks = int(impressions * (ctr / 100))
    cpc = round(np.random.uniform(0.2, 5.0), 2)
    spend = round(clicks * cpc, 2)
    cpm = round((spend / impressions) * 1000, 2)
    conversions = random.randint(0, clicks)
    conversion_rate = round((conversions / clicks) * 100 if clicks > 0 else 0, 2)
    video_views = random.randint(0, impressions)

    row = {
        'Date': random_date().strftime('%Y-%m-%d'),
        'Channel': random.choice(channels),
        'Data Source': random.choice(data_sources),
        'Campaign': random.choice(campaigns),
        'Spend': spend,
        'CPM': cpm,
        'CTR': ctr,
        'CPC': cpc,
        'Video Views': video_views,
        'Impressions': impressions,
        'Conversions': conversions,
        'Conversion Rate': conversion_rate
    }
    data.append(row)

df = pd.DataFrame(data)
```
